```python
import netCDF4 as nc
import numpy as np
import datetime
import glob
import os
import numba
#@numba.njit()

#filename='/shared_space/project_0/radar/raw/event1/2019061610100200dBR.sri_top.asc'

#Let's create the output netcdf file
file = 'dpsri_event1_test.nc'
os.system('rm %s' % file)
fp = nc.Dataset(file,'w')
#n_time=432
n_time=18
#Create dimensions
fp.createDimension("time", n_time)
fp.createDimension("lat", 799)
fp.createDimension("lon", 806)
##Create variables
#coordinates
a_lon=np.array(range(806))
a_lat=np.array(range(799))
delta_lon=a_lon*0.005669419306
delta_lat=a_lat*0.005669419306
lon=delta_lon+(-10.455520982743)
lat=delta_lat+(-37.743355940996)
#time
time= [None] * 432
for day in range(3):
    for hour in range(24):
        for ten_min in range(6):
            if hour<10:
                time[day*24*6+hour*6+ten_min]='2019'+'06'+str(16+day)+'0'+str(hour)+str(ten_min)+'0'
            else:
                time[day*24*6+hour*6+ten_min]='2019'+'06'+str(16+day)+str(hour)+str(ten_min)+'0'
ntime=len(time)                
fp.createVariable("time","i8",("time",))
fp.createVariable("lat","f8",("lat",))
fp.createVariable("lon","f8",("lon",))
fp.createVariable("precip","f4",("lat","lon","time"))

for root, dirs, files in os.walk('/shared_space/project_0/radar/raw/event1_dpsri/'):
        a=files #当前路径下所有子目录
a.sort()
nfiles = len(a)
dfiles = []
i=0
while i <= nfiles-1:
    tmp = a[i][0:12]
    dfiles.append(tmp)
    i=i+1
b=a[0:nfiles]
b.sort()
filename=b
print(len(time))
print(len(dfiles))


fp['lat'][:] = lat[:]
fp['lon'][:] = lon[:]
fp['time'][:] = time[0:n_time]
print(time[0:n_time])
print(len(time[0:n_time]))
fp['precip'].units = "mm/h" ;
fp['precip'].description = "radar rain data" ;

print('len time   ',len(time))
print('len dfiles   ',len(dfiles))
print('filename   ',len(filename))
print('ntime ', ntime)
print(len(fp['precip']))
i=0
j=0
while i <= n_time-1:
    if time[i] == dfiles[j]:
        filename_final='/shared_space/project_0/radar/raw/event1_dpsri/'+filename[j]
        print(j,'      ',filename[j],'   ', time[i], '   ', dfiles[j])
        ascii_grid = np.loadtxt(filename_final, skiprows=6)
        temp=10**(ascii_grid/10)
        fp['precip'][:,:,i] = temp[:]
        i=i+1
        j=j+1
    else:
        fp['precip'][:,:,i] = -99
        print('no data for    ',time[i])
        i=i+1
    
#Close the file
fp.close()
print('the process is finished :)')
```
